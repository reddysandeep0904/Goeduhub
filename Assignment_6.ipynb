# 4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np

x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')  # string must be enclosed with quotes '  '
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)', 'cos(x)'])      # legend entries as seperate strings in a list
plt.show()
```

# Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.



```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
  
X_axis = np.arange(len(Languages))
  
plt.bar(X_axis - 0.2, Popularity, 0.4, label = 'Popularity')
plt.bar(X_axis + 0.2, Security, 0.4, label = 'Security')
  
plt.xticks(X_axis, Languages)
plt.xlabel("Languages")
plt.ylabel("Popularity & Security Score")
plt.title("Popularity & Security of major Programming Languages")
plt.legend()
plt.show()
```

```python

plt.barh(X_axis - 0.2, Popularity, 0.4, label = 'Popularity')
plt.barh(X_axis + 0.2, Security, 0.4, label = 'Security')
  
plt.yticks(X_axis, Languages)
plt.ylabel("Languages")
plt.xlabel("Popularity & Security Score")
plt.title("Popularity & Security of major Programming Languages")
plt.legend()
plt.show()
```

# 6. Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.

Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]

Histogram showing Below Average, Average and Execellent distribution

40-60: Below Average

60-80: Average

80-100: Excellent

```python

Marks = np.array([61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
below_average = Marks[np.logical_and(Marks >= 40,Marks < 60)]
average = Marks[np.logical_and(Marks >= 60,Marks < 80)]
excellent = Marks[np.logical_and(Marks >= 80,Marks < 100)]

plt.figure(figsize=(12,6))
plt.hist(Marks, align='right', color='cyan', edgecolor='black')
plt.xlabel('Bins')
plt.ylabel('Marks')
plt.title('Histogram of Marks')
plt.vlines([40, 60, 80,100], ymin=0, ymax=4, colors='k',linestyles='dashed')
plt.show()
```

```python
print(f'No of Below Average students : {below_average.size}')
print(f'No of Average students : {average.size}')
print(f'No of Excellent students : {excellent.size}')
```
